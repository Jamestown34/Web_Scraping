# Web scraping:
is the process of automatically extracting data from websites. It involves fetching web pages, parsing the HTML or other structured formats, and extracting the desired information. Web scraping is commonly used to collect data for various purposes, including research, analysis, and automation.

Here's an overview of the basic steps involved in web scraping:

### Fetching Web Pages:
The first step in web scraping is to retrieve the HTML content of the web pages containing the data you want to extract. This is typically done using HTTP requests, either through libraries like Requests in Python or built-in functions in web scraping frameworks like Scrapy.

### Parsing HTML:
Once you have fetched the web page, you need to parse the HTML to extract the relevant data. HTML parsing libraries like BeautifulSoup (Python) or lxml (Python) are commonly used for this purpose. These libraries allow you to navigate the HTML structure, search for specific elements (e.g., <div> or <table>), and extract their contents.

### Extracting Data:
    After parsing the HTML, you can extract the desired data from the elements you've identified. This might involve retrieving text, links, images, or other types of content. You can use various methods provided by the parsing libraries to extract data efficiently.

### Cleaning and Structuring Data:
    Once you've extracted the data, you may need to clean and structure it for further processing or analysis. This could involve removing unnecessary characters, converting data types, handling missing values, and organizing the data into a structured format like lists, dictionaries, or Pandas DataFrames.

### Storing or Using the Data:
    Finally, you can choose to store the extracted data in a file (e.g., CSV, JSON, or database), display it in your application, or use it for further analysis, visualization, or automation.

It's important to note that while web scraping can be a powerful tool for accessing and utilizing data, it's essential to comply with the terms of service and legal restrictions of the websites you're scraping from. Some websites may prohibit or restrict web scraping through their terms of use or robots.txt file, so it's crucial to review and adhere to these guidelines to avoid potential legal issues.

#### data source from  Wikipedia

To see the list of  all Countries in the world


```python
import requests
from bs4 import BeautifulSoup

# Send HTTP request
url = 'https://en.wikipedia.org/wiki/List_of_countries'
response = requests.get(url)

# Parse HTML content
soup = BeautifulSoup(response.text, 'html.parser')

# Find list of countries
country_list = []
for link in soup.find_all('a'):
    if link.has_attr('title') and 'List of countries' not in link['title']:
        country_list.append(link['title'])

# Print the list of countries
print(country_list)
```

```python
import requests
from bs4 import BeautifulSoup

# URL of the Wikipedia page
url = 'https://en.wikipedia.org/wiki/Demographics_of_Nigeria'

# Send an HTTP request to the URL
response = requests.get(url)

# Parse the HTML content
soup = BeautifulSoup(response.text, 'html.parser')

# Find the table containing the data (inspect the page source to identify the table)
table = soup.find('table', class_='wikitable')

# Extract data from the table
rows = table.find_all('tr')
for row in rows:
    # Assuming each row contains data in <td> tags
    cells = row.find_all('td')
    for cell in cells:
        # Print or process the cell data as needed
        print(cell.text.strip())
```
